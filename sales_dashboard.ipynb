```python
import pandas as pd
import seaborn as sns
sns.set()

df = pd.read_csv(r"D:\Data-Analysis\Super Store Project\Sample - Superstore.csv", encoding='ISO-8859-1')
df = df.drop_duplicates()
df

```

```python
print(df.isnull().sum())
print(df.describe())
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df=df.dropna()
df
```

```python
df['Revenue']=df['Sales']
df['Revenue']=df['Sales']*df['Quantity']
df
```

```python
total_revenue = df['Revenue'].sum()
print("Total Revenue : ", round(total_revenue,2))
```

```python
#mothly sales trend
monthly_sales=df.groupby(df['Order Date'].dt.to_period('M'))['Revenue'].sum().reset_index()
monthly_sales.columns = ['Month', 'Revenue']
print(monthly_sales.head())
```

```python
#top 5 products
top_products=df.groupby('Product Name')['Revenue'].sum().sort_values(ascending=False).head(5)
print("The top 5 products\n", top_products)
```

```python
#region wise sales
region_sales=df.groupby('Region')['Revenue'].sum().sort_values(ascending=False).head(5)
print("Top 5 regions, that are doing well are\n", region_sales)
```

```python
#average order value
average_order=df.groupby('Region')["Revenue"].mean()
print("Averge order value of the order is\n", round(average_order,2))
```

```python
import matplotlib.pyplot as plt

```

```python
#monthly trend chart
plt.figure(figsize=(10,5))
monthly_sales['Month']=monthly_sales['Month'].astype(str)
plt.plot(monthly_sales['Month'], monthly_sales['Revenue'], marker='*', color='teal')
plt.title("Monthly Revenue")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_products.plot(kind='bar', figsize=(8,5), color='coral')
plt.title("TOP 5 PRODUCTS")
plt.ylabel('Revenue (₹)')
plt.xlabel('Product Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
region_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6), startangle=90)
plt.title('Sales Distribution by Region')
plt.ylabel('')  # remove y-axis label
plt.tight_layout()
plt.show()
```

```python
region_sales.plot(kind='bar', figsize=(8, 5), color='skyblue')
plt.title('Revenue by Region')
plt.ylabel('Revenue (₹)')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
with pd.ExcelWriter("Sales_Report_Styled.xlsx", engine='xlsxwriter') as writer:
    df.to_excel(writer, sheet_name='Raw Data', index=False)
    monthly_sales.to_excel(writer, sheet_name='Monthly Revenue', index=False)
    top_products.to_frame().to_excel(writer, sheet_name='Top Products')
    region_sales.to_frame().to_excel(writer, sheet_name='Region Revenue')

    # Add conditional formatting to Monthly Revenue sheet
    workbook  = writer.book
    worksheet = writer.sheets['Monthly Revenue']
    format_red = workbook.add_format({'bg_color': '#FFC7CE','font_color': '#9C0006'})
    worksheet.conditional_format('B2:B100', {'type': 'cell',
                                             'criteria': '<',
                                             'value': 10000,
                                             'format': format_red})
```
